```python
import numpy as np
import random
import sympy as sp
```

```python
#(a)T=[[2 1][3 4]]
matrix_a = np.array([[2, 1], [3, 4]])
flag=False

#Check additivity property
u=np.array([[random.randint(0,9)],[random.randint(10,20)]])
v=np.array([[random.randint(10,20)],[random.randint(0,9)]])
s=u+v

T_uv=np.dot(matrix_a,s)
T_u=np.dot(matrix_a,u)
T_v=np.dot(matrix_a,v)

if np.array_equal(T_u + T_v, T_uv):
  flag=True
else:
  flag=False

#Check homogeneity property
k=random.randint(5,15)
k_u=k*u
T_k_u=np.dot(matrix_a,k_u)
k_T_u=k*T_u
if np.array_equal(T_k_u, k_T_u):
  flag=True
else:
  flag=False

print(flag)
```

```python
x1, y1, x2, y2, c = sp.symbols('x1 y1 x2 y2 c')
u = np.array((x1, y1))
v = np.array((x2, y2))

def is_L_T(T):
  left_side = T(u+v)
  right_side = T(u)+T(v)
  additivity = np.all(left_side == right_side)
  if not additivity:
    return False

  left = T(c*u)
  right = c*T(u)
  homogenity = np.all(left==right)
  if not homogenity:
    return False

  return True
```

```python
#(b)T(x,y)=(sin(x+y),sin(x-y))
def T2(x):
  return np.array((sp.simplify(sp.sin(x[0]+x[1])),sp.simplify(sp.sin(x[0]-x[1]))))

print(is_L_T(T2))
```

```python
#(c)
def T3(x):
  return np.array((sp.simplify(x[0]*x[1]),x[1]))

print(is_L_T(T3))
```

```python
#(d)
def T4(x):
  return np.array((sp.simplify(3*x[0]-x[1]), x[1], x[0]))

print(is_L_T(T4))
```

```python
import numpy as np
import matplotlib.pyplot as plt
topx=np.array([1, .98, .8, .83, .8, .77, .5, .25, 0, -.25, -.5, -.8,
-.8, -.98, -1])
topy=np.array([0, .15, .7, 1.15, 1.03, 1.15, .85, .95, 1, .95, .85,
1.2, .7, .15, 0])
botx=-np.cos(np.pi*np.arange(1,10)/10)
boty=-np.sin(np.pi*np.arange(1,10)/10)
wiskx=np.array([.2, 1.3, .2, 1.4, .2, 1.4, .2, 1.3, .2, .17, .13, .08, .03, 0])
wisky=np.array([ 0, .3, 0, .1, 0, -.1, 0, -.3, 0, .1, -.1, .1, -.1, 0])-.2
xeye=np.array([0, .2, .3, .4, .43, .45, .43, .4, .37, .35, .37, .4, .5,
.6, .5, .4, .3, .2, 0])
yeye=np.array([.5, .5, .43, .4, .42, .5, .58, .6, .58, .5, .42, .4, .43,
.5, .57, .6, .57, .5,
.5])-.2
x=np.concatenate((topx,botx,wiskx,xeye,-xeye,-wiskx[::-1]))
y=np.concatenate((topy,boty,wisky,yeye, yeye,wisky[::-1]))
cat = np.array([x,y]).T
plt.plot(cat[:,0],cat[:,1],'b')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.show

def plot_cat(cat_coords):
    plt.plot(cat_coords[:, 0], cat_coords[:, 1], 'b')
    plt.xlim(-3, 3)
    plt.ylim(-3, 3)
    plt.show()
```

```python
#(a) Reflection through the x-axis:
#    Matrix: [[1, 0], [0, -1]]
reflection_x = np.array([[1, 0], [0, -1]])
cat_reflected_x = np.dot(cat, reflection_x)
plot_cat(cat_reflected_x)
```

```python
#(b) Reflection through the y-axis:
#    Matrix: [[-1, 0], [0, 1]]
reflection_y = np.array([[-1, 0], [0, 1]])
cat_reflected_y = np.dot(cat, reflection_y)
plot_cat(cat_reflected_y)
```

```python
#(c) Reflection through the axis y=x:
#    Matrix: [[0, 1], [1, 0]]
reflection_yx = np.array([[0, 1], [1, 0]])
cat_reflected_yx = np.dot(cat, reflection_yx)
plot_cat(cat_reflected_yx)
```

```python
#(d) Reflection through the axis y=-x:
#    Matrix: [[0, -1], [-1, 0]]
reflection_minus_yx = np.array([[0, -1], [-1, 0]])
cat_reflected_minus_yx = np.dot(cat, reflection_minus_yx)
plot_cat(cat_reflected_minus_yx)
```

```python
#(e) Reflection through the origin:
#    Matrix: [[-1, 0], [0, -1]]
reflection_origin = np.array([[-1, 0], [0, -1]])
cat_reflected_origin = np.dot(cat, reflection_origin)
plot_cat(cat_reflected_origin)
```

```python
#(f) Horizontal Contraction by a factor of 1/3:
#    Matrix: [[1/3, 0], [0, 1]]
horizontal_contraction = np.array([[1/3, 0], [0, 1]])
cat_contracted_horizontal = np.dot(cat, horizontal_contraction)
plot_cat(cat_contracted_horizontal)
```

```python
#(g) Vertical Contraction by a factor of 1/5:
#    Matrix: [[1, 0], [0, 1/5]]
vertical_contraction = np.array([[1, 0], [0, 1/5]])
cat_contracted_vertical = np.dot(cat, vertical_contraction)
plot_cat(cat_contracted_vertical)
```

```python
#(h) Horizontal Expansion by a factor of 3:
#    Matrix: [[3, 0], [0, 1]]
horizontal_expansion = np.array([[3, 0], [0, 1]])
cat_expanded_horizontal = np.dot(cat, horizontal_expansion)
plot_cat(cat_expanded_horizontal)
```

```python
#(i) Vertical Expansion by a factor of 4:
#    Matrix: [[1, 0], [0, 4]]
vertical_expansion = np.array([[1, 0], [0, 4]])
cat_expanded_vertical = np.dot(cat, vertical_expansion)
plot_cat(cat_expanded_vertical)
```

```python
#(j) Horizontal Shears by a factor of -1.5 then +1.5:
#    Matrix (first shear): [[1,-1.5], [0, 1]]
#    Matrix (second shear): [[1,1.5], [0, 1]]
horizontal_shear_left = np.array([[1, -1.5], [0, 1]])
horizontal_shear_right = np.array([[1, 1.5], [0, 1]])
cat_sheared_horizontal_left = np.dot(cat, horizontal_shear_left)
cat_sheared_horizontal_right = np.dot(cat_sheared_horizontal_left, horizontal_shear_right)
plot_cat(cat_sheared_horizontal_left)
plot_cat(cat_sheared_horizontal_right)
```

```python
#(k) Vertical Shears by a factor of -1.5 then +1.5:
#    Matrix (first shear): [[1, 0], [-1.5, 1]]
#    Matrix (second shear): [[1, 0], [1.5, 1]]
vertical_shear_down = np.array([[1, 0], [-1.5, 1]])
vertical_shear_up = np.array([[1, 0], [1.5, 1]])
cat_sheared_vertical_down = np.dot(cat, vertical_shear_down)
cat_sheared_vertical_up = np.dot(cat_sheared_vertical_down, vertical_shear_up)
plot_cat(cat_sheared_vertical_down)
plot_cat(cat_sheared_vertical_up)
```

```python
#(l) Projection onto the x-axis:
#    Matrix: [[1, 0], [0, 0]]
projection_x = np.array([[1, 0], [0, 0]])
cat_projected_x = np.dot(cat, projection_x)
plot_cat(cat_projected_x)
```

```python
#(m) Projection onto the y-axis:
#    Matrix: [[0, 0], [0, 1]]
projection_y = np.array([[0, 0], [0, 1]])
cat_projected_y = np.dot(cat, projection_y)
plot_cat(cat_projected_y)
```

```python
#(n) Rotation by an angle of /3 counterclockwise:
#    Matrix: [[cos(/3), -sin(/3)], [sin(/3), cos(/3)]]
angle_rad = np.pi/3
rotation_matrix = np.array([[np.cos(angle_rad), -np.sin(angle_rad)], [np.sin(angle_rad), np.cos(angle_rad)]])
cat_rotated_counterclockwise = np.dot(cat, rotation_matrix)
plot_cat(cat_rotated_counterclockwise)
```

```python
#(o) Rotation by an angle of /3 clockwise:
#    Matrix: [[cos(-/3), -sin(-/3)], [sin(-/3), cos(-/3)]]
angle_rad = -np.pi/3
rotation_matrix_clockwise = np.array([[np.cos(angle_rad), -np.sin(angle_rad)], [np.sin(angle_rad), np.cos(angle_rad)]])
cat_rotated_clockwise = np.matmul(cat, rotation_matrix_clockwise)
plot_cat(cat_rotated_clockwise)
```

```python
#(a)
from decimal import Decimal

# Function to perform calculations using floats
def perform_float_operations(x, y):
    addition_float = x + y
    multiplication_float = x * y
    division_float = x / y

    return addition_float, multiplication_float, division_float

# Function to perform calculations using Decimals
def perform_decimal_operations(x, y):
    x_decimal = Decimal(x)
    y_decimal = Decimal(y)

    addition_decimal = x_decimal + y_decimal
    multiplication_decimal = x_decimal * y_decimal
    division_decimal = x_decimal / y_decimal

    return addition_decimal, multiplication_decimal, division_decimal

# User input
x = float(input("Enter the first numerical value (x): "))
y = float(input("Enter the second numerical value (y): "))

# Perform calculations with float data type
float_results = perform_float_operations(x, y)

# Perform calculations with Decimal data type
decimal_results = perform_decimal_operations(x, y)

# Display results
print("\nResults:")
print("Addition (float):", float_results[0])
print("Multiplication (float):", float_results[1])
print("Division (float):", float_results[2])

print("Addition (Decimal):", decimal_results[0])
print("Multiplication (Decimal):", decimal_results[1])
print("Division (Decimal):", decimal_results[2])
```

```python
#(b)
'''Float Data Type: The float data type is appropriate for general-purpose programming and scientific calculations where a high degree of precision is not required,
or where the magnitude of numbers is not extremely large or small.
It is faster and consumes less memory compared to Decimal.

Decimal Data Type: The Decimal data type is appropriate for applications where precision is critical, such as financial calculations, scientific research involving high-precision measurements,
and situations where rounding errors can have significant consequences.
It is suitable for handling decimal numbers with a high degree of accuracy.
'''
```

```python
from math import pi
import numpy as np
import matplotlib.pyplot as plt
```

```python
#(a) To implement Leibnez formula

def pi_approx(n):
  x=0
  for i in range(n):
    x+=((-1)**i)*(1/((2*i)+1))
  return(x*4)


n=int(input())
pi_arx=pi_approx(n)
print(pi_arx)
```

```python
print(pi)
```

```python
#(b) Absolute Error between computed pi approximation and the know value of pi

def abs_error(pi_arx):
  er_abs=abs(pi_arx - pi)
  return er_abs

print(abs_error(pi_arx))
```

```python
#(c)

N_values = np.arange(1, 1001)  # Example: N values from 1 to 100
absolute_error_values = [abs_error(pi_approx(N)) for N in N_values]

# Step 3: Create the plot
plt.figure(figsize=(8, 6))
plt.plot(N_values, absolute_error_values, linestyle='-',color='red')
plt.scatter(N_values, absolute_error_values, color='blue', marker='o')
plt.title('Absolute Error vs. N')
plt.xlabel('N')
plt.ylabel('Absolute Error')
plt.xticks(np.arange(0, 1100, step=100))
plt.yticks(np.arange(0, max(absolute_error_values), step=0.05))
plt.grid(True)
plt.show()
```
